```python
# Line Chart 
# Line chart is one of the basic plots and can be created using plot() function. 
# It is used to represent a relationship between two data X and Y on a different axis.
# matplotlib.pyplot.plot(x,y)
import matplotlib.pyplot as plt
import numpy as np
house_size = [1000,1200,1500,2000,2500]
price = [100000,150000,200000,250000,300000]
plt.plot(house_size,price)
plt.title("Line Chart")
plt.xlabel("House Size")
plt.ylabel("Price")
plt.show()
```

```python
# Line chart with multiple lines
import matplotlib.pyplot as plt
import numpy as np  
house_size1 = [1000,1200,1500,2000,2500]
price1 = [100000,150000,200000,250000,300000]
house_size2 = [1000,1200,1500,2000,2500]
price2 = [120000,170000,220000,270000,320000]   
plt.plot(house_size1,price1,label="Location 1")
plt.plot(house_size2,price2,label="Location 2")
plt.title("Line Chart with Multiple Lines")
plt.xlabel("House Size")    
plt.ylabel("Price")
plt.legend()    
plt.show()
```

```python
# Line Chart with Markers
import matplotlib.pyplot as plt
import numpy as np
house_size = [1000,1200,1500,2000,2500]
price = [100000,150000,200000,250000,300000]    
plt.plot(house_size,price,marker='o')
plt.title("Line Chart with Markers")
plt.xlabel("House Size")
plt.ylabel("Price")
plt.show()
```

```python
# line chart with customized line style and color
import matplotlib.pyplot as plt
import numpy as np
house_size = [1000,1200,1500,2000,2500]
price = [100000,150000,200000,250000,300000]
plt.plot(house_size,price, color='Blue', linewidth=5,linestyle='--', marker='o')
plt.title("Customized Line Chart")
plt.xlabel("House Size")
plt.ylabel("Price")
plt.show()
```
