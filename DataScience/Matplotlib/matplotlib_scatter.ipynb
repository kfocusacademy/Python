```python
# Scatter chart examples

# Sample data
import matplotlib.pyplot as plt
cars = ['i10', 'Swift', 'Nexon', 'BMW', 'Audi']
horsepower = [68, 83, 110, 190, 240]  # X-axis
mileage = [20, 22, 18, 12, 10]        # Y-axis
scatter = plt.scatter(horsepower, mileage)

# Labels and title
plt.title('Car Horsepower vs Mileage')
plt.xlabel('Horsepower (HP)')
plt.ylabel('Mileage (km/l)')
plt.grid(True)
plt.show()
```

```python
# with formatting values using bubble size
import matplotlib.pyplot as plt
cars = ['i10', 'Swift', 'Nexon', 'BMW', 'Audi']
horsepower = [68, 83, 110, 190, 240]  # X-axis
mileage = [20, 22, 18, 12, 10]        # Y-axis
sizes = [200, 250, 300, 400, 450]     # Bubble size
colors = ['gold', 'lightgreen', 'lightcoral', 'lightskyblue', 'lightgrey']

plt.figure(figsize=(6, 4))
scatter = plt.scatter(horsepower, mileage, s=sizes, c=colors, alpha=0.7, edgecolors='black')


# Add annotations
for i, car in enumerate(cars):
    plt.text(horsepower[i]+5, mileage[i], car, fontsize=9)

# Labels and title
plt.title('Car Horsepower vs Mileage')
plt.xlabel('Horsepower (HP)')
plt.ylabel('Mileage (km/l)')
plt.grid(True)

plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Patch

# Data
height = [150, 160, 165, 170, 175, 180, 185]
weight = [50, 55, 60, 65, 70, 75, 80]
gender = ['F', 'F', 'M', 'M', 'F', 'M', 'F']
colors = ['pink' if g == 'F' else 'blue' for g in gender]

# Plot
plt.scatter(height, weight, c=colors, s=100, edgecolors='black')
plt.title('Height vs Weight by Gender')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.grid(True)

# Custom legend
legend_elements = [
    Patch(facecolor='pink', edgecolor='black', label='Female'),
    Patch(facecolor='blue', edgecolor='black', label='Male')
]
plt.legend(handles=legend_elements, title='Gender')

plt.show()
```
