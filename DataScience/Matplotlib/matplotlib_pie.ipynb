```python
# pie chart examples
import matplotlib.pyplot as plt

cars = ["i10","Swift","Nexon","BMW","Audi"]
qty = [100,130,80,10,8]
plt.pie(qty,labels=cars)
plt.title ('Cars Data')
plt.xlabel('Car model')
plt.ylabel('Quantity')
plt.show()
```

```python
# pie chart formatting examples
import matplotlib.pyplot as plt
cars = ["i10","Swift","Nexon","BMW","Audi"]
qty = [100,130,80,10,8]
colors = ['gold','lightgreen','lightcoral','lightskyblue','lightgrey']
explode = [0.1,0,0,0,0]
plt.pie(qty,labels=cars,colors=colors,explode=explode,shadow=True,startangle=140,autopct='%1.2f%%')
plt.title ('Cars Data') 
plt.xlabel('Car model')
plt.ylabel('Quantity')
plt.legend(title='Car Models')
plt.show()
```
