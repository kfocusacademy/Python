```python
# Bar graph example using matplotlib
days = ['Sun','Mon','Tue','Wed','Thu','Fri','Sat']
temp = [30,28,20,39,33,22,10]
import matplotlib.pyplot as plt
plt.bar(days,temp)
plt.title('Day wise temperature')
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.show()
```

```python
# Multiple bar graph example using matplotlib
import matplotlib.pyplot as plt
import numpy as np
days = ['Sun','Mon','Tue','Wed','Thu','Fri','Sat']
temp1 = [30,28,20,39,33,22,10]
temp2 = [25,32,34,20,25,30,15]
x1 = np.arange(len(days)) # takes values 0 to 6
x2 = [x + 0.2 for x in x1] # shift by width of bar which is 0.2
plt.bar(x1, temp1, width=0.2, label='City A', color='b')
plt.bar(x2, temp2, width=0.2, label='City B', color='r')
plt.title('Day wise temperature comparison')    
plt.xlabel('Days')
plt.ylabel('Temperature')   
plt.xticks([x + 0.1 for x in x1], days) # to display days at center of bars
plt.legend()
plt.show()
```

```python
# bar graph horizontal example using matplotlib
days = ['Sun','Mon','Tue','Wed','Thu','Fri','Sat']
temp = [30,28,20,39,33,22,10]
import matplotlib.pyplot as plt
plt.barh(days,temp)
plt.title('Day wise temperature')
plt.xlabel('Temperature')
plt.ylabel('Days')
plt.show()
```

```python
# bar graph with formatting example using matplotlib
days = ['Sun','Mon','Tue','Wed','Thu','Fri','Sat']
temp = [30,28,20,39,33,22,10]
import matplotlib.pyplot as plt 
plt.bar(days,temp, color='orange', edgecolor='black', hatch='//')
plt.title('Day wise temperature', fontsize=16, color='blue')
plt.xlabel('Days', fontsize=14)
plt.ylabel('Temperature', fontsize=14)
plt.grid(axis='y', linestyle='', alpha=0.7)
plt.show()
```
