```python
# To display bar graphs using pandas and matplotlib

import pandas as pd 
import matplotlib.pyplot as plt

df = pd.DataFrame([
    [180000, 110, 18.9, 1400],  
    [360000, 905, 23.4, 1800],  
    [230000, 230, 14.0, 1300],  
    [60000, 450, 13.5, 1500]
], columns=['Col A', 'Col B', 'Col C', 'Col D'])

print(df)
df.plot(kind='bar')
```

```python
# Normalization technique
# maximum absolute scaling - This technique rescales each feature between -1 and 1 by dividing all values by the maximum absolute value in that column
max_scaled = df.copy()

for column in max_scaled.columns:
    max_scaled[column] = max_scaled[column] / max_scaled[column].abs().max()
print(max_scaled)
max_scaled.plot(kind='bar')
```

```python
# Min max feature scaling - This technique rescales each feature to a fixed range, usually 0 to 1, by subtracting the minimum value and dividing by the range (max - min) of that column
max_scaled = df.copy()
for column in max_scaled.columns:
    max_scaled[column] = (max_scaled[column] - max_scaled[column].min()) / (max_scaled[column].max() - max_scaled[column].min())

print(max_scaled)
max_scaled.plot(kind='bar')    
```

```python
# plotting using pandas
import matplotlib.pyplot as plt
df = pd.read_csv('data_plot.csv')

# Creating a line plot
df['score'].plot(kind='line')
plt.title('Scores of Individuals')
plt.xlabel('Individual Index')
plt.ylabel('Score')
plt.show()
```

```python
# Creating bar plot of age
df.plot(kind='bar', x = 'Name', y = 'age', color = 'orange')
plt.title('Age of Individuals')
plt.xlabel('Name')  
plt.ylabel('Age')
plt.show()
```

```python
# Creating scatter plot of age vs score
df.plot(kind='scatter', x = 'age', y = 'score', color = 'green')
plt.title('Age vs Score')
plt.xlabel('Age')
plt.ylabel('Score')
plt.show()
```
