```python
# using group by to aggregate data
import pandas as pd
df = pd.read_csv('temp_city_wise.csv')

# group by city and calculate mean temperature
grouped = df.groupby('city')

for city, group in grouped:
    print(f"City: {city}")
    print(group)
    print()
```

```python
# to get specific group , max and min values 
grouped.get_group('Mumbai')
grouped.max()
grouped.min()
```

```python
# to get mean values of all cities
grouped.mean(numeric_only=True) 
```

```python
%matplotlib inline # to display plot in jupyter notebook
grouped.plot()
#import matplotlib.pyplot as plt
#grouped.mean(numeric_only=True).plot(kind='bar')
#plt.show()
```

```python
# Group by 
df = pd.read_csv('Employees.csv')

grouped = df.groupby('Dept').agg(
    Total_Salary = ('Salary','sum'),
    Avg_Salary = ('Salary','mean')
    )

print(grouped)  
```

```python
# Ranking salary within each department
df['Rank'] = df.groupby('Dept')['Salary'].rank(ascending=False)
print(df)
```
