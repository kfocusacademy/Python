```python
from abc import ABC, abstractmethod

class PaymentProcessor(ABC):
	@abstractmethod
	def process_payment(self, amount):
		pass

class phonepe(PaymentProcessor):
	def process_payment(self, amount):
		print(f"Processing payment of {amount} using PhonePe")	

	def refund_payment(self, amount):
		print(f"Refunding payment of {amount} using PhonePe")

class paytm(PaymentProcessor):
	def process_payment(self, amount):
		print(f"Processing payment of {amount} using Paytm")	

phonepe_processor = phonepe()
phonepe_processor.process_payment(1500)

phonepe_processor.refund_payment(500)

paytm_processor = paytm()
paytm_processor.process_payment(1500)
#paytm_processor.refund_payment(500)
```
