```python
# The below example is for multi inheritance 

class cars():  # Parent class
    globalname = "cars"  # class attribute 
    
    def __init__(self,carname,model):
        self.carname = carname    # Instance variable 
        self.model = model        # Instance variable 

    def startEngine(self):
        print("Start the engine",self.carname)
        
    def applyBreak(self):
        print("Applying break",self.carname)

class electriccars(cars): # Child Class1, Parent to hybrid 
    def __init__(self,carname,model,batterycapacity):
        super().__init__(carname,model)
        self.batterycapacity = batterycapacity

    def display(self):
        print(f"Car Name: {self.carname}, Model: {self.model}, Battery Capacity: {self.batterycapacity}kWh")

    def batteryCharger(self):
        print("Battery charge completed for battery capacity:",self.batterycapacity)

class hybrid(electriccars): # Inherit from both cars and electriccars 
    def __init__(self,carname,model,batterycapacity,fuelType):
        self.fuelType = fuelType
        super().__init__(carname,model,batterycapacity)
    
    def SwitchfromPetroltoEletric(self):
        print(f"Car Name: {self.carname}, Model: {self.model} Switched from petrol to electric")


alto = cars("Maruthi","Alto")
alto.applyBreak()
tesla = electriccars("Tesla","Model S",1000)
tesla.batteryCharger()
vitara = hybrid("Maruthi","Vitara",2000,"Petrol")
vitara.SwitchfromPetroltoEletric()


        
```
