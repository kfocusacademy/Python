```python
import numpy as np

no_of_test = np.array([1,2,3,4,5])
test_score = np.array([60,65,70,75,80])

# Initialize paramters
learning_rate = 0.08
iterations = 1000
n = len(no_of_test)

def gradient_descent(x,y):
    m_curr = 0
    b_curr = 0    
    for i in range(iterations):
        y_pred = m_curr * x + b_curr
        mse_cost = 1/n*sum((y - y_pred)**2)
        m_grad = -(2/n) * (sum(x * (y-y_pred)))
        b_grad = -(2/n) * (sum(y-y_pred))
        m_curr = m_curr - learning_rate * m_grad
        b_curr = b_curr - learning_rate * b_grad
        print("m{},b{},mse_cost{}, iteration{}".format(m_curr,b_curr,mse_cost,i))


gradient_descent(no_of_test,test_score)


    
```

```python
# Expected m and b using linear regression 

from sklearn import linear_model
reg = linear_model.LinearRegression()
no_of_test = np.array(no_of_test).reshape(-1,1)
reg.fit(no_of_test,test_score)
print("m{}, b{}".format(reg.coef_,reg.intercept_))

```
