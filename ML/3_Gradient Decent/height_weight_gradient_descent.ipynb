```python
# Simple gradient descent for y = mx + b
import numpy as np

# Sample data
weight = np.array([0.5, 2.3, 2.9]) 
height = np.array([1.5, 1.9, 3.2])

# Initialize parameters
m = 0
b = 0
learning_rate = 0.1
iterations = 1000
n = len(weight)

def gradient_descent(x,y):
    m_curr = 0
    b_curr = 0    
    for i in range(iterations):
        y_pred = m_curr * x + b_curr
        mse_cost = 1/n*sum((y - y_pred)**2)
        m_grad = -(2/n) * (sum(x * (y-y_pred)))
        b_grad = -(2/n) * (sum(y-y_pred))
        m_curr = m_curr - learning_rate * m_grad
        b_curr = b_curr - learning_rate * b_grad
        print("m{},b{},mse_cost{}, iteration{}".format(m_curr,b_curr,mse_cost,i))

gradient_descent(weight,height)
```
