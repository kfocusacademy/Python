```python
import pandas as pd
from sklearn import linear_model
```

```python
df = pd.read_csv("insurance.csv")
df
```

```python
# Total cost to be updated to numeric 
df["Total_cost"] = df["Total_cost"].str.replace(",","")
df["Total_cost"] = df["Total_cost"].astype(int)
df.info()
```

```python
# Linear model
df["Stay_severity"] =  df['Length_of_stay']*df['Severity']
reg = linear_model.LinearRegression()

reg.fit(df[['Stay_severity','Procedures','Age','Insurance']],df.Total_cost)
reg.predict([[8,2,25,1]])
```
