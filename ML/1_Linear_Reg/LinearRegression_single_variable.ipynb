```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
# Loading CSV with home prices 
df = pd.read_csv("homeprices2.csv")
df
```

```python
# Drawing scatter plot with area vs price
plt.xlabel('Area (SFT)')
plt.ylabel('Price $')
plt.scatter(df.area, df.price, color = 'red',marker="+")
```

```python
# With linear regression line
%matplotlib inline
plt.xlabel('Area (SFT)',fontsize =20)
plt.ylabel('Price $',fontsize = 20)
plt.scatter(df.area, df.price, color = 'red',marker = '+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
reg.predict ([[5500]])
```

```python
# With linear regression line
%matplotlib inline
plt.xlabel('Area (SFT)',fontsize =20)
plt.ylabel('Price $',fontsize = 20)
plt.scatter(df.area, df.price, color = 'red',marker = '+')
plt.plot(df.area,reg.predict(df[['area']]),color= 'blue')
```

```python
print('Coef (or) m ', reg.coef_)
print('Intercept (or) b',reg.intercept_)
```
