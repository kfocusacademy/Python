```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
# Load dataset
df = pd.read_csv('ice_cream_sales.csv')
df
```

```python
# Draw a scatter chart
plt.xlabel = "Temp(C)"
plt.ylabel = "Ice Cream Sales"
plt.scatter(df.temp,df.Ice_cream_sales,color = "Red", marker= "+")
plt.show()
```

```python
# Perform Linear Regression Prediction
reg = linear_model.LinearRegression()
# Temp should be passed it as 2D array, ice cream sales as 1D array
reg.fit(df[['temp']],df['Ice_cream_sales']) 
reg.predict([[28]])
```

```python
print(df[['temp']].shape)           
print(df['Ice_cream_sales'].shape) # Should be (n_samples,)
```

```python
# To update sales prediction for new data
df2 = pd.read_csv('ice_cream_sales_test.csv')
df2
df2['Predicted_sales'] = reg.predict(df2[['temp']])
df2.to_csv('ice_cream_sales_test_with_predictions.csv', index=False)
```
